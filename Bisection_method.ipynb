# Bisection Method

The goal of a root finding method is to solve the equation f(x) = 0 or in other words, find the roots of an equation. The bisection method is one such root finding method to find the roots of a function. It is called a bracketing method because it involves 'bracketing' the solution with an initial upper and lower guess.

In these examples I will be using the simple equation 2 = x^3. The first step is defining your function so that all known and unknown values are on one side of the equation and 0 is on the other.

For example: 2 = x^3 becomes 0 = x^3 - 2

x^3 - 2 is now in the form f(x) = 0 and can be solved using a root finding method.

The next step is to perfom the actual bisection method, which I have tried to make into steps list for convinience.

(1) Check to see if your brackets include a sign change

(2) Apply the actual bisection method

(3) Check again for a sign change and reassign your midpoint to be your new upper or lower bracket guess

A root findind method is an iterative process which means that you need some sort of cutoff criteria for it. There are 2 types of cutoff critera we use in this class; maximum iteration count and convergence. The first example is of maximum iteration, the second example is of convergence, and the third example is using both.

```python
# Example 1

# Function to find the roots of
def func(x):
    return x**3 - 2

def bisect(func, upper, lower, maxit=100):
    counter = 0
    prev = 100

    # step 1
    if func(upper)*func(lower) > 0:
        return "Error, inital guesses do not bracket answer"

    for i in range(maxit):
        # step 2
        midpoint = (upper+lower)/2

        # step 3
        if func(midpoint)*func(upper) > 0:
            upper = midpoint
        else:
            lower = midpoint

        counter += 1

    return midpoint,counter

sol, iterations = bisect(func=func, upper=52, lower=-62)
print(f"Solution: {sol}, Number of Iterations: {iterations}")
```

As you can see from example 1, the code went through all 100 iterations. This is good, but lets look at the second cuttof condition, convergence.

```python
# Example 2

# Function to find the roots of
def func(x):
    return x**3 - 2

def bisect(func, upper, lower, maxit=100):
    counter = 0
    prev = 100

    # step 1
    if func(upper)*func(lower) > 0:
        return "Error, inital guesses do not bracket answer"

    while True:
        # step 2
        midpoint = (upper+lower)/2

        # Convergence cutoff critera
        ea = abs((midpoint - prev) / midpoint) * 100
        prev = midpoint
        if ea < 0.1:
            break

        # step 3
        if func(midpoint)*func(upper) > 0:
            upper = midpoint
        else:
            lower = midpoint

        counter += 1

    return midpoint,counter

sol, iterations = bisect(func=func, upper=52, lower=-62)
print(f"Solution: {sol}, Number of Iterations: {iterations}")
```

In this example the code only ran through 16 iterations! This means the bracketing method was able to reach an error of 0.1% in only 16 iterations. In example 1 we ran through 84 extra iterations. These extra iterations may have given a more accurate answer but an error of 0.1% is usually good enough for most engineering appplications.

```python
# Example 3

# Function to find the roots of
def func(x):
    return x**3 - 2

def bisect(func, upper, lower, maxit=100):
    counter = 0
    prev = 100

    # Step 1
    if func(upper)*func(lower) > 0:
        return "Error, inital guesses do not bracket answer"

    for i in range(maxit):
        # step 2
        midpoint = (upper+lower)/2

        ea = abs((midpoint - prev) / midpoint) * 100
        prev = midpoint
        if ea < 0.1:
            break

        # Step 3
        if func(midpoint)*func(upper) > 0:
            upper = midpoint
        else:
            lower = midpoint

        counter += 1

    return midpoint,counter

sol, iterations = bisect(func=func, upper=52, lower=-62)
print(f"Solution: {sol}, Number of Iterations: {iterations}")
```

As you can see examples 2 and 3 returned the same answer. So you may be asking yourself, "Why would we want to use maximum iterations as a cutoff criteria?" Good question. The benefit to using the maximum iterations criteria is that it acts as a fail safe in case your solution does not converge. If your solution starts to diverge then it will never meet the convergence critera and go on forever. But if you include the maximum iteration criteria then even if your solutions diverge you will exit the loop when the maximum number of iterations is met.

It is worth noting that the bracketing method will always converge (provided the function is continuious and the root exists). But open root finding methods are not garunteed to converge.
