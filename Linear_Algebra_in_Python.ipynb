#Linear Algebra

The purpose of this page is to give an overview of how to solve A*x = b using python. A simple understanding of matrices is assumed (for example what a NxM matrix means). However, an understanding of system conditioning is not assumed.

```python
import numpy as np

A = np.array([[1,2],[4,5]])
print(A)
A_t = A.T
print(A_t)
A_inv = np.linalg.inv(A)
print(A_inv)
```
